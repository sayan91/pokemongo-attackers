```python
# Adding the weakness and resistance information to dictionaries based on table above. 
# The resistances and weaknesses are the 4th and 5th columns respectively.

weakness = {}


weakness["Normal"] = ["Fighting"]
weakness["Fighting"] = ["Flying","Psychic","Fairy"]
weakness["Flying"] = ["Electric","Ice","Rock"]
weakness["Poison"] = ["Ground","Psychic"]
weakness["Ground"] = ["Water","Ice","Grass"]
weakness["Rock"] = ["Grass","Fighting","Water","Steel","Ground"]
weakness["Bug"] = ["Flying","Fire","Rock"]
weakness["Ghost"] = ["Ghost","Dark"]
weakness["Steel"] = ["Fighting","Ground","Fire"]
weakness["Fire"] = ["Water","Ground","Rock"]
weakness["Water"] = ["Grass","Electric"]
weakness["Grass"] = ["Flying","Fire","Ice","Bug","Poison"]
weakness["Electric"] = ["Ground"]
weakness["Psychic"] = ["Bug","Ghost","Dark"]
weakness["Ice"] = ["Fire","Rock","Steel","Fighting"]
weakness["Dragon"] = ["Dragon","Ice","Fairy"]
weakness["Fairy"] = ["Steel","Poison"]
weakness["nan"] = []


resistance = {}

resistance["Normal"] = ["Ghost"]
resistance["Fighting"] = ["Dark","Bug","Rock"]
resistance["Flying"] = ["Fighting","Ground","Grass"]
resistance["Poison"] = ["Poison","Grass","Fairy"]
resistance["Ground"] = ["Poison","Electric","Rock"]
resistance["Rock"] = ["Normal","Poison","Fire","Flying"]
resistance["Bug"] = ["Fighting","Ground","Grass"]
resistance["Ghost"] = ["Normal","Fighting","Poison","Bug"]
resistance["Steel"] = ["Normal","Flying","Poison","Rock","Bug","Steel","Grass","Psychic","Ice","Dragon","Fairy"]
resistance["Water"] = ["Water","Steel","Ice","Fire"]
resistance["Fire"] = ["Bug","Steel","Fire","Grass","Ice"]
resistance["Grass"] = ["Water","Ground","Grass","Electric"]
resistance["Electric"] = ["Electric","Steel","Flying"]
resistance["Psychic"] = ["Fighting","Psychic"]
resistance["Ice"] = ["Ice"]
resistance["Dragon"] = ["Water","Fire","Grass","Electric"]
resistance["Fairy"] = ["Fighting","Bug","Dragon","Dark"]
resistance["nan"] = []

legendary_pokemon = ["Mewtwo", "Mew", "Zapdos", "Moltres", "Articuno"]
```

```python
zipped_list = zip(list(raw_df["Pokemon"]),list(raw_df["P_Type_1"]) ,list(raw_df["P_Type_2"]))
pokemon_types = {}
for pokemon, type1, type2 in zipped_list:
    if pokemon not in pokemon_types.keys():
        pokemon_types[pokemon] = (type1, str(type2))
        
# pokemon_types
                                                
```

```python
def move_effectivity(move_type, pokemon):
    RESISTANCE_MULT = 0.8
    WEAKNESS_MULT = 1.25
    
    pokemon_weaknesses = weakness[pokemon_types[pokemon][0]] + weakness[pokemon_types[pokemon][1]]
    weakness_count = pokemon_weaknesses.count(move_type)
    pokemon_resistances = resistance[pokemon_types[pokemon][0]] + resistance[pokemon_types[pokemon][1]]
    resistance_count = pokemon_resistances.count(move_type)
    
    effectivity = weakness_count - resistance_count

    if effectivity > 0:
        return WEAKNESS_MULT**abs(effectivity)
    else:
        return RESISTANCE_MULT**abs(effectivity)
```
